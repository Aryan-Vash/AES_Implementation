# AES-128 Encryption Notebook
This notebook encrypts one 128-bit block, shows matrix states after every AES step, computes a 6-character polynomial rolling hash, and stores ciphertext in ciphertext.txt.

```python
from pathlib import Path
import json

from aes_key_expansion import expand_key
from aes_state import bytes_to_state, state_to_bytes
from aes_mix_columns import mix_columns
from aes_sbox import sub_bytes
from aes_shift_rows import shift_rows
from aes_xor import add_round_key
from hash import polynomial_rolling_hash


def matrix_to_hex_rows(state):
    return [" ".join(f"{value:02x}" for value in row) for row in state]


def write_text(path, value):
    with open(path, "w", encoding="utf-8") as file:
        file.write(value)


def write_json(path, value):
    with open(path, "w", encoding="utf-8") as file:
        json.dump(value, file, indent=2)


plaintext_path = Path('plaintext.txt')
key_path = Path('key.txt')

plaintext = plaintext_path.read_text(encoding='utf-8').strip() if plaintext_path.exists() else 'Two One Nine Two'
key_hex = key_path.read_text(encoding='utf-8').strip() if key_path.exists() else '5468617473206d79204b756e67204675'

if len(plaintext.encode('utf-8')) != 16:
    raise ValueError('Plaintext must be exactly 16 bytes in UTF-8.')
if len(key_hex) != 32:
    raise ValueError('Key in key.txt must be exactly 32 hex characters (16 bytes).')

plaintext_bytes = plaintext.encode('utf-8')
try:
    key_bytes = bytes.fromhex(key_hex)
except ValueError as error:
    raise ValueError('Key in key.txt must contain valid hexadecimal characters.') from error


def show_state(label, state):
    print(label)
    for row in matrix_to_hex_rows(state):
        print('  ' + row)
    print()


print('Plaintext:', plaintext)
print('Key (hex):', key_hex)
```

```python
# returns keys for all 11 round keys (round 0 to round 10) in nested list form
round_keys=expand_key(key_bytes)

# converts 16-byte plaintext into AES 4x4 state matrix
state = bytes_to_state(plaintext_bytes)

# stores states in JSON so encryption and decryption steps can be compared later
encryption_trace = {
    'initial_plaintext_state': [row[:] for row in state],
    'after_initial_add_round_key': None,
    'rounds': []
}

# initial AddRoundKey (round 0): XOR plaintext state with first round key
show_state('Initial Plaintext State', state)
state = add_round_key(state, round_keys[0])
encryption_trace['after_initial_add_round_key'] = [row[:] for row in state]
show_state('After Initial AddRoundKey', state)
```

```python
round_number = 1
round_trace = {'round': round_number}

#from here till round 9 we will see the 4 same repeated operations again and again
#individual code of the function is in their individual files

state = sub_bytes(state)
round_trace['after_sub_bytes'] = [row[:] for row in state]
show_state('Round 1 - After SubBytes', state)

state = shift_rows(state)
round_trace['after_shift_rows'] = [row[:] for row in state]
show_state('Round 1 - After ShiftRows', state)

state = mix_columns(state)
round_trace['after_mix_columns'] = [row[:] for row in state]
show_state('Round 1 - After MixColumns', state)

state = add_round_key(state, round_keys[round_number])
round_trace['after_add_round_key'] = [row[:] for row in state]
show_state('Round 1 - After AddRoundKey', state)

encryption_trace['rounds'].append(round_trace)
```

```python
round_number = 10
round_trace = {'round': round_number}

# final round: same as other rounds, but MixColumns is skipped
state = sub_bytes(state)
round_trace['after_sub_bytes'] = [row[:] for row in state]
show_state('Round 10 - After SubBytes', state)

state = shift_rows(state)
round_trace['after_shift_rows'] = [row[:] for row in state]
show_state('Round 10 - After ShiftRows', state)

state = add_round_key(state, round_keys[round_number])
round_trace['after_add_round_key'] = [row[:] for row in state]
show_state('Round 10 - After AddRoundKey', state)

encryption_trace['rounds'].append(round_trace)

ciphertext_bytes = state_to_bytes(state)
ciphertext_hex = ciphertext_bytes.hex()

encryption_trace['ciphertext_state'] = [row[:] for row in state]
encryption_trace['ciphertext_hex'] = ciphertext_hex
print('Ciphertext (hex):', ciphertext_hex)
```

```python
rolling_hash_6 = polynomial_rolling_hash(ciphertext_hex, width=6)
print('6-character Polynomial Rolling Hash:', rolling_hash_6)

#add hash to cyphertext and save the results
ciphertext_payload = f"{ciphertext_hex}|{rolling_hash_6}"
ciphertext_utf = ciphertext_bytes.decode('utf-8', errors='replace')
plaintext_hex = plaintext_bytes.hex()
key_hex = key_bytes.hex()
input_output_hex_pair = f"input_hex: {plaintext_hex}\noutput_hex: {ciphertext_hex}\nkey_hex: {key_hex}"

write_text('ciphertext.txt', ciphertext_payload)
write_text('ciphertext_utf.txt', ciphertext_utf)
write_text('plaintext.txt', plaintext)
write_text('key.txt', key_hex)
write_text('input_output_hex.txt', input_output_hex_pair)
write_json('encryption_trace.json', encryption_trace)

print('Saved ciphertext+hash (hex payload) to ciphertext.txt')
print('Saved UTF-converted ciphertext to ciphertext_utf.txt')
print('Saved plaintext to plaintext.txt')
print('Saved 32-char hex key to key.txt')
print('Saved input/output/key hex trio to input_output_hex.txt')
print('Saved full encryption trace to encryption_trace.json')
```
